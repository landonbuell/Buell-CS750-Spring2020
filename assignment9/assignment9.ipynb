# Problem 2 [50%] 
### Fit a small neural network model to the ‘ISLR::Caravan” dataset (You can apparently).

```python
# Load in Caravan data
caravan = pd.read_csv("Caravan.csv",index_col=0)
caravan.head(10)
```

```python
# break into feature matrix & target vector
y = caravan['Purchase'].replace(['No','Yes'],[0,1])
X = caravan.drop('Purchase',axis=1)

# dimensionality reduction:
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
X_scaled = scale(X)
pca = PCA(n_components=20)
pca = pca.fit(X=X_scaled)
X = pca.transform(X_scaled)

# print shapes!
print(X.shape)
print(y.shape)

# train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4)

```

```python
# Create & Train Neural Network (20,20,20,20)

MLP_CLF = MLPClassifier(hidden_layer_sizes=(20,20,20,20),activation='relu',
                       solver='sgd',max_iter=400,tol=1e-4,
                       random_state=0)

MLP_CLF = MLP_CLF.fit(X_train,y_train)

# test the model w/ Confusion Matrix

y_pred = MLP_CLF.predict(X_test)
confmat = confusion_matrix(y_test,y_pred)
print(confmat)
```

```python
# Create & Train Neural Network (100,)

MLP_CLF = MLPClassifier(hidden_layer_sizes=(100,),activation='relu',
                       solver='sgd',max_iter=400,tol=1e-4,
                       random_state=1)

MLP_CLF = MLP_CLF.fit(X_train,y_train)

# test the model w/ Confusion Matrix

y_pred = MLP_CLF.predict(X_test)
confmat = confusion_matrix(y_test,y_pred)
print(confmat)
```
