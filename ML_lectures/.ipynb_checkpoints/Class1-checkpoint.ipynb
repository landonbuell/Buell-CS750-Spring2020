```python
# From John's Notebook, load in Dataset

from sklearn.datasets import load_boston
dataset = load_boston()
X,y = dataset['data'],dataset['target']

print(dataset['DESCR'])
```

```python
# Split into Training & Testing Dataset

from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3)

print("Training Dimesnions:",xtrain.shape,ytrain.shape)
print("Testing Dimesnions:",xtest.shape,ytest.shape)
```

```python
# Use some linear regression

from sklearn.linear_model import LinearRegression

lin_model = LinearRegression()
lin_model.fit(xtrain,ytrain)

predictions = lin_model.predict(xtest)

# score of the model retunrs R^2 score 
# See Landon's note for R^2 score
lin_model.score(xtest,ytest)
```

```python
import numpy as np

# variance & std devof output
var = ((predictions - ytest)**2).mean()
sd = np.sqrt(var)

import matplotlib.pyplot as plt
plt.plot(ytest,predictions,'ob')
plt.fill_between([0,50],[0,50]+sd,[0,50]-sd,alpha=0.3,label='$\sigma$')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# LOOK into this!
forest = RandomForestRegressor()

forest.fit(xtrain,ytrain)
forest.score(xtest,ytest)
```
